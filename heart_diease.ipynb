```python
# Handling missing values
num_cols = ['cigsPerDay', 'BPMeds', 'totChol', 'BMI', 'heartRate','glucose','education']
data[num_cols] = data[num_cols].fillna(data[num_cols].median())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Data = StandardScaler()


Z_cols = ['totChol','sysBP','diaBP','BMI','heartRate','cigsPerDay']

# Apply standardization to the specified columns
data[Z_cols] =Data.fit_transform(data[Z_cols])
```

```python
Y=data['TenYearCHD']
data.drop('TenYearCHD',axis=1,inplace=True)
X=data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.3,random_state=42)
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)


```

```python
model.fit(X_train, Y_train)

```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
```

```python
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))

print("\nf1 Score (weighted):", f1_score(Y_test, y_pred, average='weighted'))


```

```python
import pickle
with open("heart_disease_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model saved successfully!")
```
